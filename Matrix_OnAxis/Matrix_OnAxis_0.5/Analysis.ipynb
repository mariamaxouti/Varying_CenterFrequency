```python
## plots the results
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

frequency_list = np.loadtxt('freq_data_matrix_array_with_window_without_SciFi.csv', delimiter = ',')      # [Hz]
frequency_list_MHz = [f/1000000 for f in frequency_list]   # [MHz]
amplitude_list = np.loadtxt('amplitude_data_matrix_array_with_window_without_SciFi.csv', delimiter = ',')

plt.plot(frequency_list_MHz, amplitude_list, linestyle='-',lw = 2.0)


def calculate_frequency_range(center_frequency_mhz, bandwidth_fraction):
    # Convert center frequency to Hertz
    center_frequency_hz = center_frequency_mhz * 1e6

    # Calculate the bandwidth in Hertz
    bandwidth_hz = bandwidth_fraction * center_frequency_hz

    # Calculate the minimum and maximum frequencies
    min_frequency_hz = center_frequency_hz - (bandwidth_hz / 2)
    max_frequency_hz = center_frequency_hz + (bandwidth_hz / 2)

    # Convert frequencies back to MHz for the output
    min_frequency_mhz = min_frequency_hz / 1e6
    max_frequency_mhz = max_frequency_hz / 1e6

    return min_frequency_mhz, max_frequency_mhz

center_frequency_mhz = 3.5  # Replace with your center frequency in MHz
bandwidth_fraction = 0.6     # Replace with your bandwidth fraction (e.g., 0.6 for 60%)

f_min, f_max = calculate_frequency_range(center_frequency_mhz, bandwidth_fraction)

# Draw the faint yellow square
highlight_rect = patches.Rectangle((f_min, 0), f_max - f_min, max(amplitude_list), linewidth=0.5, edgecolor='yellow', facecolor='yellow', alpha=0.5)
plt.gca().add_patch(highlight_rect)
plt.text((f_min + f_max) / 2, 5, 'Transducer Bandwidth', color='black', ha='center', va='center', fontsize=16)

plt.title('Amplitude-Frequency Spectrum (matrix array - on-axis)',fontsize=26)
plt.xlabel('Frequency [MHz]',fontsize=24)
plt.ylabel('Amplitude',fontsize = 24)
plt.xlim([0,5])
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
fig.savefig('amplitude_frequency_spectum__matrix_array_with_window_without_SciFi_bandwidth.png',transparent = False, dpi = 300, facecolor='white');
```
