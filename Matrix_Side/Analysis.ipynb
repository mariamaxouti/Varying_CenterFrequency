```python
plt.style.use('default')

# plots the results
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

voxel_size = 0.1 # [mm]
nVoxels = 155

depthList = np.arange(0,nVoxels)
#depthList = np.linspace(-nVoxels/2,nVoxels/2,nVoxels)

# loads the pressure distribution data along the z-axis
source_p0_x = np.loadtxt('int_p0_x_source.csv', delimiter = ',')
plt.plot(depthList*voxel_size, source_p0_x, linestyle='-', color='black', lw = 3.0, zorder=0,label='Source')

# uploads the reconstructed pressure distribution (along z-axis) files - produced from k-Wave
recon_files = ['matrix_array_int_p0_x_rec_1_with_window_without_SciFi.csv', 
               'matrix_array_int_p0_x_rec_2_with_window_without_SciFi.csv', 
               'matrix_array_int_p0_x_rec_3_with_window_without_SciFi.csv']
 
files = len(recon_files)
for i in range(files):
    recon_data = np.loadtxt(recon_files[i], delimiter = ',')
    if i == 0:
        iter = 1
    if i == 1:
        iter = 2
    if i == 2:
        iter = 3
    # plots the data
    plt.plot(depthList*voxel_size, recon_data, linestyle='--', lw = 3.0, zorder=0,label='R'+str(iter)) 
plt.legend(fontsize=18, frameon=True, facecolor='white', framealpha=1, edgecolor='black')

plt.title(f'Pressure distribution along x-axis', fontsize=26)
plt.xlabel('x-position [mm]', fontsize=24)
plt.ylabel('Pressure [Pa]', fontsize = 24)
#plt.xlim([-6,6])
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
fig.savefig(f'distribution_iter_with_window_without_SciFi_x.png',transparent = False, dpi = 300, facecolor='white');
```
